```python
from IPython.core.display import HTML
from IPython.display import Image
import pandas as pd
import os, sys
import re

cell_path = 'Individual_cells'
```

```python
all_the_cells = os.listdir(cell_path)

pruned = []
unpruned = []
overlays = []
time_list = []
id_list = []

for cell in all_the_cells:
    time = re.findall(r'\d+', cell)
    if time != []:
        time = int(time[0])
        im_path = cell_path + '/'+cell + '/masks'
        for mask_name in os.listdir(im_path):
            if mask_name[len(mask_name)-4:len(mask_name)] != '.png':
                continue
            ID = int(mask_name[5:6])
            pruned.append(cell_path + '/' + cell + '/skeletons' + '/pruned_' + str(ID) +'.png')
            unpruned.append(cell_path + '/' + cell + '/skeletons' + '/unpruned_' + str(ID)+'.png')
            overlays.append(cell_path + '/' + cell + '/skeletons' + '/overlay_' + str(ID) +'.png')
            time_list.append(time)
            id_list.append(ID)
```

```python
outdf=pd.DataFrame()
outdf['Pruned']=pruned
outdf['Unpruned'] = unpruned
outdf['Overlay'] = overlays
outdf['times'] = time_list
outdf['IDs'] = id_list
outdf = outdf.sort_values(by='times').reset_index()
def out_formatter(path):
    return '<img src="'+ path + '" width="256" >'
```

```python
imagehtml = outdf.to_html(escape=False,formatters=dict(Pruned=out_formatter,Unpruned=out_formatter,Overlay=out_formatter))
HTML (imagehtml)
```
